<div style="display: grid; grid-template-columns: repeat(auto-fit, minmax(250px, 1fr)); gap: 20px; margin: 25px 0;">

<!-- AND Operator -->
<div style="background: white; border-radius: 12px; padding: 20px; text-align: center; box-shadow: 0 6px 20px rgba(0,0,0,0.1); border: 3px solid #00b894;">
<div style="font-size: 2.5em; margin-bottom: 10px;">🤝</div>
<h3 style="color: #00b894; margin: 0 0 10px 0;">AND</h3>
<p style="color: #2d3436; margin: 0 0 15px 0;">True only if <strong>both</strong> are True</p>
<div style="background: #f8f9fa; padding: 10px; border-radius: 6px; font-family: monospace; font-size: 0.9em;">

True AND False = <span style="color: #e17055; font-weight: bold;">False</span>
</div>
</div>

<!-- OR Operator -->
<div style="background: white; border-radius: 12px; padding: 20px; text-align: center; box-shadow: 0 6px 20px rgba(0,0,0,0.1); border: 3px solid #fdcb6e;">
<div style="font-size: 2.5em; margin-bottom: 10px;">🔗</div>
<h3 style="color: #e17055; margin: 0 0 10px 0;">OR</h3>
<p style="color: #2d3436; margin: 0 0 15px 0;">True if <strong>at least one</strong> is True</p>
<div style="background: #f8f9fa; padding: 10px; border-radius: 6px; font-family: monospace; font-size: 0.9em;">
True OR False = <span style="color: #00b894; font-weight: bold;">True</span><br>
False OR False = <span style="color: #e17055; font-weight: bold;">False</span>
</div>
</div>

<!-- NOT Operator -->
<div style="background: white; border-radius: 12px; padding: 20px; text-align: center; box-shadow: 0 6px 20px rgba(0,0,0,0.1); border: 3px solid #a29bfe;">
<div style="font-size: 2.5em; margin-bottom: 10px;">🔄</div>
<h3 style="color: #6c5ce7; margin: 0 0 10px 0;">NOT</h3>
<p style="color: #2d3436; margin: 0 0 15px 0;">Flips True to False and vice versa</p>
<div style="background: #f8f9fa; padding: 10px; border-radius: 6px; font-family: monospace; font-size: 0.9em;">
NOT True = <span style="color: #e17055; font-weight: bold;">False</span><br>
NOT False = <span style="color: #00b894; font-weight: bold;">True</span>

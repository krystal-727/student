```python
%%html

<div id="outputPH1">Results will appear here:<br></div>
<h3>1. Mood Check</h3>
<input id="moodInput" placeholder = "Enter your mood (happy, mad, sad, excited)">
<button onclick="checkMood()">Check Mood</button>
<div id="moodOutput">Results will appear here:<br></div>

<script>


   function checkMood() {
    const mood = document.getElementById("moodInput").value.toLowerCase();
    const output = document.getElementById("moodOutput");

    if (mood === "happy") {
        output.innerText += "You're happy! Here's some ice cream!\n";
    } else if (mood === "mad") {
        output.innerText += "You're mad. Take a deep breath.\n";
    } else if (mood === "sad") {
        output.innerText += "You're sad. Sending virtual hugs.\n";
    } else if (mood === "excited") {
        output.innerText += "You're excited! Let's celebrate!\n";
    } else {
        output.innerText += "Mood undefined. Try happy, mad, sad, or excited.\n";
    }
}
</script> 
    
  
    <h3>2. Weather Check</h3>
<input id="weatherInput" placeholder="Enter today's weather (sunny, rainy, cloudy, snowy)">
<button onclick="checkWeather()">Check Weather</button>
<div id="weatherOutput">Weather result:<br></div>

 
<script> 
  function checkWeather() {
    const weather = document.getElementById("weatherInput").value.toLowerCase();
    const output = document.getElementById("weatherOutput");
 
    if (weather === "sunny") {
        output.innerText += "It's sunny outside!\n";
    } else if (weather === "cloudy") {
        output.innerText += "It's cloudy outside.\n";
    } else if (weather === "rainy") {
        output.innerText += "It's rainy outside.\n";
    } else if (weather === "snowy") {
        output.innerText += "It's snowy outside.\n";
    } else {
        output.innerText += "Weather not recognized.\n";
    }
}
</script>
</script>
<h3>3. Age Checker</h3>
<button onclick="checkAge()">Generate Random Age and Check</button>
<div id="ageOutput">Age results:<br></div>

<script>
function checkAge() {
    const output = document.getElementById("ageOutput");
    const age = Math.floor(Math.random() * 100); // Random age between 0-99

    if (age < 13) {
        output.innerText += `Age ${age}: You're a child.\n`;
    } else if (age >= 13 && age < 18) {
        output.innerText += `Age ${age}: You're a teenager.\n`;
    } else if (age >= 18 && age < 65) {
        output.innerText += `Age ${age}: You're an adult.\n`;
    } else {
        output.innerText += `Age ${age}: You're a senior citizen.\n`;
    }
}
</script>

```
